```python
#Guessing number game and login window using tkinter
import tkinter as tk
import random
from tkinter import messagebox
 
def login(event=None):
    username=Username_entry.get()
    password=Password_entry.get()

    if username=="admin" and password=="password":
        messagebox.showinfo(title="Login Successfull",message="Welcome back!")
        login_window.destroy() # to close the login window
        guess_number_game() # to start the game
    else:
        messagebox.showerror(title="Login Failed",message="Invalid username or password")

def guess_number_game():
    
    # creating a game window
    game_window=tk.Tk()
    game_window.title("Guess the number game")
    game_window.geometry("840x440")
    game_window.configure(bg="#333333")

    # generating a random number
    number = random.randint(1,10)

    def check_guess(event=None):
        try:
            guess=int(guess_entry.get())
            if guess<number:
                feedback_label.config(text="Too low! Try again.")
            elif guess>number:
                feedback_label.config(text="Too high! Try again.")
            else:
                feedback_label.config(text="Congratulations! You guessed it!")
                guess_button.config(state=tk.DISABLED)
        except ValueError:
            feedback_label.config(text="Please enter a valid number.")

    #creating and placing widget for game window
    instructions=tk.Label(game_window,text="Guess a number between 1 to 10",bg="#333333",fg="#ff3333",font=("Rockwell",24,"italic"))
    instructions.pack(pady=20)
    guess_entry=tk.Entry(game_window,font=("Rockwell",14,"italic"))
    guess_entry.pack(pady=10)
    guess_button=tk.Button(game_window,text="Submit",font=("Rockwell",14),bg="#333333",fg="#ff3399",command=check_guess)
    guess_button.pack(pady=10)
    feedback_label=tk.Label(game_window,text="",bg="#333333",fg="#ff1131",font=("Arial",14))
    feedback_label.pack(pady=10)

    game_window.bind("<Return>",check_guess) 

# creating a main window for login
login_window=tk.Tk()
login_window.title("Login Page")
login_window.geometry("840x440")
login_window.configure(bg='#333333')

#putting all widget in a frame
frame=tk.Frame(bg="#333333")

#Creating Widget for login page
login_Label=tk.Label(frame,text="Login",bg="#333333",fg="#ff3333",font=("Arial",30))
Username_Label=tk.Label(frame,text="Username",bg="#333333",fg="#ffffff",font=("Arial",16))
Username_entry=tk.Entry(frame,font=("Arial",16))
Password_Label=tk.Label(frame,text="Password",bg="#333333",fg="#ffffff",font=("Arial",16))
Password_entry=tk.Entry(frame,show="*",font=("Arial",16))
Login_button=tk.Button(frame,text="Login",bg="#333333",fg="#ff3333",font=("Arial",16),command=login)

#Placing Widget of login page on the screen
login_Label.grid(row=0, column=0, columnspan=2,pady=40)
Username_Label.grid(row=1,column=0)
Username_entry.grid(row=1,column=1,pady=20)
Password_Label.grid(row=2,column=0)
Password_entry.grid(row=2,column=1,pady=20)
Login_button.grid(row=3,column=0,columnspan=2,pady=40)

Login_button.bind("<Return>",login)

frame.pack()
login_window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox, filedialog
from fpdf import FPDF
import datetime
import os

# Shop Details
SHOP_NAME = "Maa Vindywashini Hardware"
SHOP_ADDRESS = "Bediban Madhuban Chouk, Pipra, East Champaran, Bihar"
SHOP_PHONE = "6200411069"
SHOP_EMAIL = "maavindywashinihardware@gmail.com"

# Initialize global variables
cart = []
GST_RATE = 18  # Default GST rate
user_details = {"name": "", "gst_number": ""}
bill_counter = 1  # Start with the first bill

# Function to generate a new bill number
def generate_bill_number():
    global bill_counter
    return f"MVH/24-25/{bill_counter:03d}"

# Function to set user details
def set_user_details():
    user_details["name"] = entry_name.get()
    user_details["gst_number"] = entry_gst.get()
    gst_rate_window.destroy()

# Function to select GST rate
def select_gst_rate():
    global GST_RATE
    GST_RATE = gst_rate_var.get()
    messagebox.showinfo("GST Rate Updated", f"GST rate set to {GST_RATE}%.")

# Function to add item to the cart
def add_item():
    item_name = entry_item_name.get()
    try:
        price = float(entry_price.get())
        quantity = int(entry_quantity.get())
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid price and quantity.")
        return

    if item_name and price > 0 and quantity > 0:
        cart.append({"name": item_name, "price": price, "quantity": quantity})
        update_cart()
        entry_item_name.delete(0, tk.END)
        entry_price.delete(0, tk.END)
        entry_quantity.delete(0, tk.END)
    else:
        messagebox.showerror("Input Error", "Please enter valid item details.")

# Function to update the cart display
def update_cart():
    listbox_cart.delete(0, tk.END)
    for item in cart:
        listbox_cart.insert(tk.END, f"{item['name']} - ₹{item['price']} x {item['quantity']}")

# Function to generate the bill
def generate_bill():
    if not cart:
        messagebox.showerror("Cart Empty", "Add items to the cart before generating the bill.")
        return

    global bill_counter
    bill_number = generate_bill_number()
    total = 0
    bill_details = f"{'Item':<20}{'Qty':<10}{'Price':<10}{'Total':<10}\n"
    bill_details += "-" * 50 + "\n"

    for item in cart:
        item_total = item['price'] * item['quantity']
        total += item_total
        bill_details += f"{item['name']:<20}{item['quantity']:<10}₹{item['price']:<10}₹{item_total:<10}\n"

    gst_amount = total * GST_RATE / 100
    grand_total = total + gst_amount

    bill_details += "-" * 50 + "\n"
    bill_details += f"{'Subtotal':<40}₹{total:.2f}\n"
    bill_details += f"{'GST ({GST_RATE}%)':<40}₹{gst_amount:.2f}\n"
    bill_details += f"{'Grand Total':<40}₹{grand_total:.2f}\n"

    # Display the bill
    show_bill(bill_number, bill_details, total, gst_amount, grand_total)

    # Increment bill counter
    bill_counter += 1

# Function to display the bill
def show_bill(bill_number, bill_details, total, gst_amount, grand_total):
    bill_window = tk.Toplevel(root)
    bill_window.title("Bill Summary")

    # Add shop details
    tk.Label(bill_window, text=SHOP_NAME, font=("Arial", 14, "bold")).pack()
    tk.Label(bill_window, text=SHOP_ADDRESS, font=("Arial", 10)).pack()
    tk.Label(bill_window, text=f"Phone: {SHOP_PHONE} | Email: {SHOP_EMAIL}", font=("Arial", 10)).pack()
    tk.Label(bill_window, text="-" * 50).pack()

    # Add bill details
    bill_text = tk.Text(bill_window, wrap=tk.WORD, width=50, height=20)
    bill_text.insert(tk.END, f"Bill Number: {bill_number}\n")
    bill_text.insert(tk.END, f"Date: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    bill_text.insert(tk.END, f"Customer Name: {user_details['name']}\n")
    bill_text.insert(tk.END, f"GST Number: {user_details['gst_number']}\n\n")
    bill_text.insert(tk.END, bill_details)
    bill_text.insert(tk.END, f"\nGrand Total: ₹{grand_total:.2f}")
    bill_text.config(state=tk.DISABLED)
    bill_text.pack(padx=10, pady=10)

    # Add buttons
    tk.Button(bill_window, text="Save as PDF", command=lambda: save_bill_as_pdf(bill_number, bill_details, total, gst_amount, grand_total)).pack(pady=5)
    tk.Button(bill_window, text="Close", command=bill_window.destroy).pack(pady=5)

# Function to save bill as a PDF
def save_bill_as_pdf(bill_number, bill_details, total, gst_amount, grand_total):
    pdf = FPDF()
    pdf.add_page()

    # Use Arial font, built-in for FPDF
    pdf.set_font('Arial', size=12)

    # Shop details
    pdf.cell(200, 10, txt=SHOP_NAME, ln=True, align='C')
    pdf.cell(200, 10, txt=SHOP_ADDRESS, ln=True, align='C')
    pdf.cell(200, 10, txt=f"Phone: {SHOP_PHONE} | Email: {SHOP_EMAIL}", ln=True, align='C')
    pdf.cell(200, 10, txt=" ", ln=True)

    # Bill header
    pdf.cell(100, 10, txt=f"Bill Number: {bill_number}", ln=True)
    pdf.cell(100, 10, txt=f"Date: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", ln=True)
    pdf.cell(100, 10, txt=f"Customer Name: {user_details['name']}", ln=True)
    pdf.cell(100, 10, txt=f"GST Number: {user_details['gst_number']}", ln=True)
    pdf.cell(200, 10, txt=" ", ln=True)

    # Bill details
    pdf.multi_cell(0, 10, txt=bill_details)

    # Grand Total
    pdf.set_font('Arial', size=12, style='B')
    pdf.cell(100, 10, txt=f"Grand Total: ₹{grand_total:.2f}", ln=True)

    # Save PDF
    file_path = filedialog.asksaveasfilename(
        defaultextension=".pdf",
        filetypes=[("PDF files", "*.pdf")],
        initialfile=f"{bill_number}.pdf"
    )
    if not file_path:  # Check if user canceled the dialog
        messagebox.showinfo("Save Canceled", "The bill was not saved as a PDF.")
        return

    try:
        pdf.output(file_path)
        messagebox.showinfo("Success", f"Bill saved as {file_path}")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to save the PDF: {e}")

# Main Window
root = tk.Tk()
root.title("POS Billing System with GST")

# User Details Frame
frame_user = tk.Frame(root, padx=10, pady=10)
frame_user.pack()

tk.Label(frame_user, text="Customer Name:").grid(row=0, column=0, padx=5, pady=5)
entry_name = tk.Entry(frame_user)
entry_name.grid(row=0, column=1, padx=5, pady=5)

tk.Label(frame_user, text="GST Number:").grid(row=1, column=0, padx=5, pady=5)
entry_gst = tk.Entry(frame_user)
entry_gst.grid(row=1, column=1, padx=5, pady=5)

tk.Button(frame_user, text="Set Details", command=set_user_details).grid(row=2, column=0, columnspan=2, pady=10)

# Item Entry Frame
frame_entry = tk.Frame(root, padx=10, pady=10)
frame_entry.pack()

tk.Label(frame_entry, text="Item Name:").grid(row=0, column=0, padx=5, pady=5)
entry_item_name = tk.Entry(frame_entry)
entry_item_name.grid(row=0, column=1, padx=5, pady=5)

tk.Label(frame_entry, text="Price (₹):").grid(row=1, column=0, padx=5, pady=5)
entry_price = tk.Entry(frame_entry)
entry_price.grid(row=1, column=1, padx=5, pady=5)

tk.Label(frame_entry, text="Quantity:").grid(row=2, column=0, padx=5, pady=5)
entry_quantity = tk.Entry(frame_entry)
entry_quantity.grid(row=2, column=1, padx=5, pady=5)

tk.Button(frame_entry, text="Add Item", command=add_item).grid(row=3, column=0, columnspan=2, pady=10)

# Cart Frame
frame_cart = tk.Frame(root, padx=10, pady=10)
frame_cart.pack()

tk.Label(frame_cart, text="Cart Items:").pack()
listbox_cart = tk.Listbox(frame_cart, width=50, height=10)
listbox_cart.pack()

# GST Rate Selection
gst_rate_var = tk.IntVar(value=GST_RATE)
frame_gst = tk.Frame(root, padx=10, pady=10)
frame_gst.pack()

tk.Label(frame_gst, text="Select GST Rate:").grid(row=0, column=0, padx=5, pady=5)
tk.Radiobutton(frame_gst, text="5%", variable=gst_rate_var, value=5, command=select_gst_rate).grid(row=0, column=1)
tk.Radiobutton(frame_gst, text="12%", variable=gst_rate_var, value=12, command=select_gst_rate).grid(row=0, column=2)
tk.Radiobutton(frame_gst, text="18%", variable=gst_rate_var, value=18, command=select_gst_rate).grid(row=0, column=3)
tk.Radiobutton(frame_gst, text="28%", variable=gst_rate_var, value=28, command=select_gst_rate).grid(row=0, column=4)

# Buttons
frame_buttons = tk.Frame(root, padx=10, pady=10)
frame_buttons.pack()

tk.Button(frame_buttons, text="Generate Bill", command=generate_bill).grid(row=0, column=0, padx=10)
tk.Button(frame_buttons, text="Exit", command=root.quit).grid(row=0, column=1, padx=10)

root.mainloop()
```
